```python
import pandas as pd
import numpy as np

from datetime import datetime
from collections import Counter
```

```python
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv('data/18 - sources of income_cleaned.csv')
```

```python
def p2f(x):
    return float(x.strip('%'))
```

```python
private_df = df.loc[df.funding_type == 'Private income']
nums = private_df[private_df.columns[-13:]]
nums_clean = nums.replace('%','',regex=True).astype('float')
res = nums_clean.mean(axis=1)
average_out = res[:-1]
national_average = res.iloc[-1]
average_out.sort()
labels = list(df.institution[average_out.index].values)
average_out.index = labels
x_national_average = np.ones(len(labels))*national_average
```

```python
aspect_mult = 1.0
figsize(aspect_mult*16,aspect_mult*9)

linewidth = 3
fig = pyplot.figure()
ax = fig.add_subplot(111) 
average_out.plot(kind='bar',label="Individual Institution")

ax.set_title("Average Private income as % of institution funding 2000-2012")
ax.set_ylabel("% of institution funding")
ax.set_xlabel("Institution")

ax.plot(x_national_average,color="r",label="National Average")
plt.legend(loc="upper left")
plt.text(0.5, -0.255,'(Data Source: CHET || Author: Vukosi Marivate (http://www.vima.co.za / @vukosi)',
     horizontalalignment='right',
#      verticalalignment='center',
     transform = ax.transAxes,color='black',fontsize=12)
pyplot.savefig('plots/private_income.png',bbox_inches='tight', dpi = 300)
```
